```python
import pandas as pd
```

```python
df1= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\Andhrapradesh_bus_details.csv")
df2= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\bihar_bus_details.csv")
df3= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\gujarat_bus_details.csv")
df4= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\haryana_bus_details.csv")
df5= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\kerala_bus_details.csv")
df6= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\punjab_bus_details.csv")
df7= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\rajasthan_bus_details.csv")
df8= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\southbengal_bus_details.csv")
df9= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\telangana_bus_details.csv")
df10= pd.read_csv(r"C:\Users\praba\Desktop\red_bus_project\bus_details\uttarpradesh_bus_details.csv")

```

```python
df_combined = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], ignore_index=True)

```

```python
df_combined
```

```python
df_combined.info()
```

```python
df_combined.isnull().sum()
```

```python
#Replace NaN with a string Value and numeric value into float
df_combined['Ratings'] = df_combined['Ratings'].apply(lambda x: float(x.split('\n')[0]) if isinstance(x, str) and '\n' in x else x)  
df_combined['Ratings'] = df_combined['Ratings'].fillna("No Rating")

```

```python
df_combined["Ratings"].unique()
```

```python
# Replace empty strings and "New" with "No Rating"
df_combined['Ratings'] = df_combined['Ratings'].replace([' ', 'New'], 'No Rating')
```

```python
df_combined["Ratings"].unique()
```

```python
df_combined.isnull().sum()
```

```python
# Convert Start_time and End_time to 12-hour format
df_combined['Start_time'] = pd.to_datetime(df_combined['Start_time'], format='%H:%M').dt.strftime('%I:%M %p')
df_combined['End_time'] = pd.to_datetime(df_combined['End_time'], format='%H:%M').dt.strftime('%I:%M %p')
```

```python
# Remove "seats available" and convert to integer
df_combined['Seats_Available'] = df_combined['Seats_Available'].str.extract('(\d+)').astype(int)
```

```python
import pymysql
import pandas as pd

# Your MySQL server credentials
host = "localhost"  # or your server IP
user = "root"
password = "praba12"
database = "redbus_scrap"

# Connect to the MySQL server
connection = pymysql.connect(host=host, user=user, password=password, database=database)
cursor = connection.cursor()

# Define the table creation query
create_table_query = """
CREATE TABLE IF NOT EXISTS bus_details (
    Bus_name VARCHAR(255),
    Bus_type VARCHAR(255),
    Start_time VARCHAR(50),
    End_time VARCHAR(50),
    Total_duration VARCHAR(50),
    Price FLOAT,
    Seats_Available INT,
    Ratings VARCHAR(50),
    Route_name VARCHAR(255),
    Route_link TEXT
);
"""
try:
    # Create the table
    cursor.execute(create_table_query)
    print("Table 'bus_details' created successfully.")
except Exception as e:
    print("Error while creating table:", e)

# Insert rows from dataframe into the MySQL table
try:
    # Insert query template
    insert_query = """
    INSERT INTO bus_details (Bus_name, Bus_type, Start_time, End_time, Total_duration, Price, Seats_Available, Ratings, Route_name, Route_link)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
    """
    
    # Iterate through the dataframe and insert rows
    for _, row in df_combined.iterrows():
        cursor.execute(insert_query, tuple(row))
    
    # Commit the changes
    connection.commit()
    print("Data inserted successfully into 'bus_details'.")
except Exception as e:
    print("Error while inserting data:", e)
finally:
    # Close the connection
    cursor.close()
    connection.close()
```
